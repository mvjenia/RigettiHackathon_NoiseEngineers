```python
from pyquil.quil import Program
from pyquil.api import QVMConnection
from pyquil.gates import *
from pyquil.api import CompilerConnection
import numpy as np
from math import pi
```

```python
from pyquil.api import QPUConnection
from pyquil.quil import Pragma
qpu = QPUConnection('19Q-Acorn')
```

```python
num_flips = 1000
angl=pi/8.0

for k in range(16):
    for j in range(16):

        h0=1.0*k
        omegat=1.0*j
        print([k,j])
        
        p = Program()
        p.inst(Pragma("PRESERVE_BLOCK"))
        for i in range(299): #Here half of the above is the allowed # of gates
            p.inst(RZ((omegat+h0)*angl, 0))    
            p.inst(X(0))
            p.inst(RZ((omegat-h0)*angl, 0))
            p.inst(X(0))
        p.inst(Pragma("END_BLOCK"))
        p.measure(0, 0)
        results =qpu.run(p, [0], num_flips)
        print(results[0:5])
        print(np.sum(results)/num_flips)

        p = Program()
        p.inst(Pragma("PRESERVE_BLOCK"))
        p.inst(X(0))
        for i in range(299):
            p.inst(RZ((omegat+h0)*angl, 0))    
            p.inst(X(0))
            p.inst(RZ((omegat-h0)*angl, 0))
            p.inst(X(0))
        p.inst(Pragma("END_BLOCK"))
        p.measure(0, 0)
        results =qpu.run(p, [0], num_flips)
        print(results[0:5])
        print(np.sum(results)/num_flips)
```

```python
deltapn = np.array(deltap)
```

It turns out that the data is well explained by sin(4pi h0/16)cos(4pi omegat/16)

```python
fitarr =np.sin(4*pi*deltapn[:,0]/16)*np.cos(4*pi*deltapn[:,1]/16)
distinct =np.unique(fitarr)
distinct =np.unique(np.round(distinct,2))
distinct
```

```python
z= 0*distinct
inde= 0
for y in distinct:
    meandeltap =0.0
    counterr =0
    for i in range(np.size(fitarr)):
        if np.round(fitarr,2)[i]==y:
            meandeltap= meandeltap+deltapn[i,2]
            counterr=counterr+1
            
    meandeltap=meandeltap/counterr
    print(meandeltap)
    z[inde]=meandeltap
    inde=inde+1
```

```python
print(distinct)
print(z)
```

Let's add some errorbars.

```python
errbar= 0*distinct
inde= 0
for y in distinct:
    
    
    
    errbar[inde]=np.std(deltapn[np.round(fitarr,2)==y,2])/np.sqrt(np.sum(np.round(fitarr,2)==y))
    inde=inde+1
```
